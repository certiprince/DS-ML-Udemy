# K means clustering

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,
                 random_state=101)
```

```python
# tuple
data
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
# if had no labels for data, then would be done here already
# in our case, we have the labels from data[1]
# kmeans clustering not used to predict labels, as it is unsupervised
# only used to try to assign already existing labels
kmeans.labels_
```

```python
fig , (ax1,ax2) = plt.subplots(1,2, sharey=True,figsize=[10,6])

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
